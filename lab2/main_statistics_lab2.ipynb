# Импорт библиотек

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats
from scipy.stats import variation
from statistics import mode
from lab3_utils import bernoulli, exponential, central_stats, variabiliyy_stats, distribution_shape_stats, plot_distributions, robustness_analysis
```

# Генерация двух распределений

```python
data_binom = bernoulli(n=10, p=0.3, size=1000)
data_exp = exponential(scale=2.0, size=1000)
```

# Анализ биномиального распределения

```python
central_stats(data_binom, "Биномиальное")
variabiliyy_stats(data_binom, "Биномиальное")
distribution_shape_stats(data_binom, "Биномиальное")
plot_distributions(
    data_binom, 
    "Биномиальное", 
    "discrete", 
    {'n': 10, 'p': 0.3}
)
```

# Анализ экспоненциального распределения

```python
central_stats(data_exp, "Экспоненциальное")
variabiliyy_stats(data_exp, "Экспоненциальное")
distribution_shape_stats(data_exp, "Экспоненциальное")
plot_distributions(
    data_exp, 
    "Экспоненциальное", 
    "continuous", 
    {'scale': 2.0}  # Параметры распределения
)
```

# Анализ устойчивости

```python
robustness_analysis(
    data_exp,
    "Экспоненциальное",
    "exponential",
    {'scale': 2.0}
)
```
